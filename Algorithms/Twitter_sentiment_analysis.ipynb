```python
import pandas as pd
datapath = "Datasets/twitter_dataset/twitter_training.csv"
df = pd.read_csv(datapath)
df.shape
```

```python
df.columns = ['tweet_id','entity','sentiment','tweet_content']
```

lowercase

```python
df['tweet_content'] = df["tweet_content"].str.lower()
```

```python
df['tweet_content'][7]
```

remove html tags

```python
import re
def remove_html_tags(text):
    pattern = re.compile("<.*?>")
    return pattern.sub(r" ",text)
```

```python
df['tweet_content'] = df["tweet_content"].apply(remove_html_tags)
```

spelling correction

```python
from textblob import TextBlob
def get_polarity(text):
    if pd.isna(text): # Handle NaN/missing values
        return None
    return TextBlob(str(text)).sentiment.polarity

# Function to get the subjectivity score (0.0 to 1.0)
def get_subjectivity(text):
    if pd.isna(text):
        return None
    return TextBlob(str(text)).sentiment.subjectivity
```

```python
df['polarity']=df['tweet_content'].apply(get_polarity)
df['subjectivity'] = df['tweet_content'].apply(get_subjectivity)
```

```python
df.head()
```

```python
def correct_tweet_spelling(tweet):
    if isinstance(tweet,str):
        return str(TextBlob(tweet).correct())
    return tweet
df['corrected_tweet_content'] = df['tweet_content'].apply(correct_tweet_spelling)
```
